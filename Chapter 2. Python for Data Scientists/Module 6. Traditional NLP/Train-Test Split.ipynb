```python
import pandas as pd

df = pd.read_csv("../DATA/twitter_sentiment_train.csv")
df
```

```python
df = df.drop(columns="id")
df
```

```python
remove_idx = df[df["label"] == 0][:10000]
```

```python
df = df.drop(remove_idx.index)
```

```python
df
```

```python
df = df.sample(frac=1).reset_index(drop=True)
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.3)
```

```python
train
```

```python
train.to_csv('../DATA/twitter_sentiment_train_.csv', columns=["label", "tweet"], index=False)
test.to_csv('../DATA/twitter_sentiment_test_.csv', columns=["label", "tweet"], index=False)  
```
