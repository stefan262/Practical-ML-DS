```python
import pandas as pd

path = '../Data/Teaching/synthetic'

df = pd.read_fwf(path,header = None)

df.head()
```

```python
df_main = pd.DataFrame()


for column in df:
    if column != 9:
        df_main = pd.concat([df_main, df[column]],axis = 0, ignore_index = True)



df_main.head()
```
